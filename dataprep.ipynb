#### **B.9 Mengatasi outlier menggunakan metode IQR**

```python
def IQR(data):
    Q3 = data.quantile(0.75)
    Q1 = data.quantile(0.25)
    IQR = Q3 - Q1
    
    upper = Q3 + 1.5 * IQR
    lower = Q1 - 1.5 * IQR
    
    return upper, lower
```

```python
for col in df.columns:
    upper, lower = IQR(df[col])
    df[col] = df[col].apply(lambda x: np.percentile(df[col], 95) if x > upper or x < lower else x)
```

#### **C.4 Mengkonfigurasi random state untuk melakukan OVERSAMPLING**

#### **C.5 Menetapkan variabel Dependen dan Independen**

```python
X = df.iloc[:, 1:]
y = df.iloc[:, 0]
```

#### **C.6 Mengkonfigurasi parameter yang dibutuhkan untuk melakukan OVERSAMPLING**

```python
labels, counts = np.unique(y, return_counts=True)
n_sample = np.max(counts) - np.min(counts)
minority = labels[counts.argmin()]
```

#### **C.7 Melakukan FEATURE SCALING pada dataset**

```python
from metrics import StandardScaler
X_scaled = StandardScaler(X)
```

#### **C.8 Melakukan OVERSAMPLING pada dataset**

```python
from metrics import Oversampled

sampled = Oversampled(random_state, n_sample=n_sample)
X_resampled, y_resampled = sampled.resampled(X_scaled, y, minority)
```

#### **C.9 Melakukan SPLITTING DATA training, dan testing pada variabel dependen dan independen**

```python
from metrics import train_test_split

train_size = 0.8
X_train, X_test, y_train, y_test = train_test_split(X=X_resampled, y=y_resampled, train_size=train_size, random_state=random_state, stratisfy=y)
```
